```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from gensim.models import KeyedVectors
from sklearn.metrics import average_precision_score
from sklearn.model_selection import StratifiedKFold
```

```python
class AdsPairDataset(Dataset):
    def __init__(self, df, ft, num_feats, max_len=50):
        self.df = df.reset_index(drop=True)
        self.ft = ft
        self.num_feats = num_feats
        self.max_len = max_len

    def __len__(self):
        return len(self.df)

    def embed_text(self, title, desc):
        tokens = (title + ' ' + desc).split()[:self.max_len]
        vecs = [self.ft[t] for t in tokens if t in self.ft.key_to_index]
        if not vecs:
            return np.zeros(self.ft.vector_size, dtype=np.float32)
        return np.mean(vecs, axis=0)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        v1 = torch.from_numpy(self.embed_text(row['base_title'], row['base_description']))
        v2 = torch.from_numpy(self.embed_text(row['cand_title'], row['cand_description']))
        n1 = torch.tensor([row[f] for f in self.num_feats], dtype=torch.float32)
        n2 = torch.tensor([row[f] for f in self.num_feats], dtype=torch.float32)
        label = torch.tensor(row.get('is_double', 0), dtype=torch.float32)
        return v1, v2, n1, n2, label
```

```python
class ContrastiveModel(nn.Module):
    def __init__(self, emb_dim, hidden_dim, num_feat_dim):
        super().__init__()
        total = emb_dim*2 + num_feat_dim*2
        self.fc = nn.Sequential(
            nn.Linear(total, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, hidden_dim)
        )
    def forward(self, v1, v2, n1, n2):
        x = torch.cat([v1, v2, n1, n2], dim=1)
        z = self.fc(x)
        return z
```

```python
def contrastive_loss(z, label, margin=1.0):
    dim = z.size(1)//2
    z1, z2 = z[:, :dim], z[:, dim:]
    dist = torch.norm(z1-z2, p=2, dim=1)
    pos = label * dist.pow(2)
    neg = (1-label) * torch.clamp(margin-dist, min=0).pow(2)
    return (pos+neg).mean()
```

```python
def train_epoch(model, loader, opt, device):
    model.train(); total=0
    for v1,v2,n1,n2,lbl in loader:
        v1,v2,n1,n2,lbl = [x.to(device) for x in (v1,v2,n1,n2,lbl)]
        opt.zero_grad()
        out = model(v1,v2,n1,n2)
        loss = contrastive_loss(out,lbl)
        loss.backward(); opt.step()
        total += loss.item()*v1.size(0)
    return total/len(loader.dataset)
```

```python
def get_preds(model, loader, device):
    model.eval()
    sims=[]
    with torch.no_grad():
        for v1,v2,n1,n2,_ in loader:
            v1,v2,n1,n2=[x.to(device) for x in (v1,v2,n1,n2)]
            out=model(v1,v2,n1,n2)
            dim=out.size(1)//2
            z1,z2=out[:,:dim],out[:,dim:]
            dist=torch.norm(z1-z2,p=2,dim=1).cpu().numpy()
            sims.extend((1.0/(1.0+dist)).tolist())
    return np.array(sims)
```

```python
def evaluate_map(model, loader, device):
    model.eval()
    labels, preds = [], []
    with torch.no_grad():
        for v1, v2, n1, n2, lbl in loader:
            v1 = v1.to(device, non_blocking=True)
            v2 = v2.to(device, non_blocking=True)
            n1 = n1.to(device, non_blocking=True)
            n2 = n2.to(device, non_blocking=True)
            out = model(v1, v2, n1, n2)
            half = out.size(1) // 2
            z1, z2 = out[:, :half], out[:, half:]
            dist = torch.norm(z1 - z2, p=2, dim=1).cpu().numpy()
            sim = 1.0 / (1.0 + dist)
            labels.extend(lbl.cpu().numpy())
            preds.extend(sim)
    return average_precision_score(labels, preds)
```

```python
ft = KeyedVectors.load_word2vec_format('cc.ru.300.vec', binary=False, unicode_errors='ignore', limit=100000)
emb_dim=ft.vector_size
```

```python
num_feats=['cand_price', 'base_price', 'cand_count_images', 'base_count_images']
```

```python
skf=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
all_test_preds = np.zeros(len(data_test))
val_scores = []
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
torch.backends.cudnn.benchmark = True
```

```python
for fold,(tr,va) in enumerate(skf.split(data_train, data_train['is_double'])):
    print(f"Fold {fold+1}")
    train_df = data_train.iloc[tr]
    val_df   = data_train.iloc[va]

    train_ds = AdsPairDataset(train_df, ft, num_feats, max_len=30)
    val_ds   = AdsPairDataset(val_df,   ft, num_feats, max_len=30)
    test_ds  = AdsPairDataset(data_test,ft, num_feats, max_len=30)

    weights = [1.0/train_df['is_double'].value_counts()[l] for l in train_df['is_double']]
    sampler = WeightedRandomSampler(weights, len(weights), replacement=True)

    train_loader = DataLoader(train_ds, batch_size=128, sampler=sampler, num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=128, shuffle=False, num_workers=1, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=128, shuffle=False, num_workers=1, pin_memory=True)

    model = ContrastiveModel(vector_dim, 128, len(num_feats)).to(device)
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    for epoch in range(1,5):
        loss = train_epoch(model, train_loader, optimizer, device)
        print(f"  Epoch {epoch}, Loss={loss:.4f}")
    val_preds = get_preds(model, val_loader, device)
    score = average_precision_score(val_df['is_double'], val_preds)
    val_scores.append(score)
    print(f"  Fold {fold+1} mAP: {score:.4f}\n")

    test_preds = get_preds(model, test_loader, device)
    all_test_preds += test_preds

print(f"Avg CV mAP: {np.mean(val_scores):.4f}")
```

```python
all_test_preds /= skf.n_splits
submission=pd.DataFrame({
    'base_id':data_test['base_item_id'],
    'cand_id':data_test['cand_item_id'],
    'probability':all_test_preds
})
submission.to_csv('submission.csv',index=False)
```

```python
import numpy as np
import pandas as pd
from gensim.models import KeyedVectors
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.metrics import average_precision_score
```

```python
!wget -nc https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.ru.300.vec.gz
!gunzip -f cc.ru.300.vec.gz
```

```python
for col in ['base_title','cand_title','base_description','cand_description']:
    data_train[col].fillna('', inplace=True)
    data_test[col].fillna('', inplace=True)
```

```python
ft = KeyedVectors.load_word2vec_format(
    'cc.ru.300.vec', binary=False, unicode_errors='ignore', limit=20000
)
```

```python
def embed_pair(row, ft, max_len=20):
    def mean_emb(prefix):
        tokens = (row[f'{prefix}_title'] + ' ' + row[f'{prefix}_description']).split()[:max_len]
        vecs = [ft[w] for w in tokens if w in ft.key_to_index]
        return np.mean(vecs, axis=0) if vecs else np.zeros(ft.vector_size)
    b, c = mean_emb('base'), mean_emb('cand')
    return np.hstack([np.abs(b-c), b*c])
```

```python
y = data_train['is_double'].values
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from sklearn.utils.class_weight import compute_class_weight

classes = np.array([0, 1])
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
weight_dict = dict(zip(classes, class_weights))
```

```python
clf = SGDClassifier(loss='log_loss', class_weight=weight_dict)
```

```python
all_test_preds = np.zeros(len(data_test))
val_scores = []
classes = np.array([0,1])
for fold, (tr_idx, va_idx) in enumerate(skf.split(data_train, y)):
    print(f"Fold {fold+1}")
    for start in range(0, len(tr_idx), 10000):
        batch_idx = tr_idx[start:start+10000]
        X_batch = np.vstack([embed_pair(data_train.iloc[i], ft) for i in batch_idx])
        y_batch = y[batch_idx]
        if start == 0:
            clf.partial_fit(X_batch, y_batch, classes=classes)
        else:
            clf.partial_fit(X_batch, y_batch)
    X_val = np.vstack([embed_pair(data_train.iloc[i], ft) for i in va_idx])
    y_val = y[va_idx]
    val_pred = clf.predict_proba(X_val)[:,1]
    score = average_precision_score(y_val, val_pred)
    val_scores.append(score)
    print(f"  mAP: {score:.4f}")
    preds = []
    for start in range(0, len(data_test), 10000):
        batch = data_test.iloc[start:start+10000]
        X_test_batch = np.vstack([embed_pair(row, ft) for _, row in batch.iterrows()])
        preds.extend(clf.predict_proba(X_test_batch)[:,1])
    all_test_preds += np.array(preds)

print(f"Mean CV mAP: {np.mean(val_scores):.4f}")
```

```python
all_test_preds /= skf.n_splits
submission = pd.DataFrame({
    'base_id': data_test['base_item_id'],
    'cand_id': data_test['cand_item_id'],
    'probability': all_test_preds
})
submission.to_csv('submission.csv', index=False)
```
